```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("supply_chain_data.csv")
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df["Total_Delivery_Time"] = (df["Order_Processing_Time"] + df["Lead_Time_Days"] + df["Shipping_Time_Days"])
```

```python
df["Delivery_Delay_Flag"] = df["Total_Delivery_Time"].apply(lambda x: "Normal" if x > 15 else "Delayed")
```

```python
df["Stock_to_Demand_Ratio"] = (df["Current_Stock"] / df["Demand_Forecast"])
```

```python
df["Logistics_Cost"] = (df["Transportation_Cost"] +df["Storage_Cost"] +df["Operational_Cost"])
```

```python
df["Delivery_Efficiency"] = (df["Monthly_Sales"] / df["Total_Delivery_Time"])
```

```python
def stock_category(ratio):
    if ratio < 0.9:
        return "Shortage"
    elif ratio <= 1.1:
        return "Balanced"
    else:
        return "Excess"

df["Inventory_Status"] = df["Stock_to_Demand_Ratio"].apply(stock_category)
```

```python
df[[
"Total_Delivery_Time",
"Delivery_Delay_Flag",
"Stock_to_Demand_Ratio",
"Logistics_Cost",
"Delivery_Efficiency",
"Inventory_Status"
]].head()
```

```python
df['Total_Delivery_Time'] = df['Total_Delivery_Time'].round(0).astype('int64')
```

```python
df['Stock_to_Demand_Ratio'] = df['Stock_to_Demand_Ratio'].round(2).astype('float64')
```

```python
df['Logistics_Cost'] = df['Logistics_Cost'].round(2).astype('float64')
```

```python
df['Delivery_Efficiency'] = df['Delivery_Efficiency'].round(2).astype('int64')
```

```python
df['Order_Processing_Time'] = df['Order_Processing_Time'].round(2).astype('float64')
```

```python
df['Return_Rate'] = df['Return_Rate'].round(2).astype('float64')
```

```python
df['Customer_Rating'] = df['Customer_Rating'].round(2).astype('float64')
```

```python
df.to_csv("Cleaned_supply_chain.csv", index=False)
```
